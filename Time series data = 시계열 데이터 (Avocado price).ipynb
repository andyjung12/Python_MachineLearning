```python
import numpy as np
import pandas as pd
from fbprophet import Prophet
```

```python
df = pd.read_csv('./avocado.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.groupby('type').mean()
```

```python
# date, average price, region=TotalUS
df = df.loc[(df.type == 'conventional') & (df.region == 'TotalUS')]
df['Date'] = pd.to_datetime(df['Date'])
data = df[['Date', 'AveragePrice']].reset_index(drop=True)
data = data.rename(columns = {'Date':'ds', 'AveragePrice':'y'})
data.head()
```

```python
data.plot(x = 'ds', y = 'y', figsize=(16,8))
```

```python
model = Prophet()
model.fit(data)
future = model.make_future_dataframe(periods = 365)
forecast = model.predict(future)
forecast.tail()
```

```python
#미래 예측 2018/05 이후
fig1 = model.plot(forecast)
#검은색 점 = 아보카도 가격
```

```python
fig2 = model.plot_components(forecast)
#트랜드 예측
#연간 가격 예측
```
